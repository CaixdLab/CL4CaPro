```python
import numpy as np
import matplotlib.pyplot as plt

# Separate the data into different lists for easier plotting
cancer_types = [item[0] for item in data_list]
uncensored_counts = [item[1] for item in data_list]
censored_counts = [item[2] for item in data_list]

# Set the positions for the bars
positions = np.arange(len(cancer_types))

# Create the bar plot
fig, ax = plt.subplots(figsize=(3, 6))

fig.patch.set_facecolor('none')  # This sets the background of the figure to transparent
ax.patch.set_facecolor('none')  # This sets the background of the subplot to transparent

# Remove background grid
ax.grid(False)

# Plot uncensored counts
bars_uncensored = ax.barh(positions, uncensored_counts, color='#FFA07A', label='High Risk')

# Plot censored counts to the right of uncensored counts
bars_censored = ax.barh(positions, censored_counts, left=uncensored_counts, color='#D8BFD8', label='Low Risk')

# Describe the data
ax.tick_params(axis='x', labelsize=7)
ax.tick_params(axis='y', labelsize=7)
ax.set_yticks(positions, fontsize=7)
ax.set_yticklabels(cancer_types, fontsize=7)
ax.set_xlabel('Number of patients', fontsize=7)
ax.set_ylabel('Cancer Type', fontsize=7)
ax.legend(fontsize=7)

max_total_count = 0  # To store the maximum total count

# Annotate bars with the sample counts
for pos, uncen, cen in zip(positions, uncensored_counts, censored_counts):
    total_count = '(' + str(uncen) + '/' + str(cen) + ')'
    total_count_num = uncen + cen
    if total_count_num > 0:
        ax.text(total_count_num + 0.5, pos, total_count, va='center', fontsize=7)
    max_total_count = max(max_total_count, total_count_num)  # Update the maximum total count

# Adjust x-axis limit to ensure text fits
ax.set_xlim([0, max_total_count + 120])  # Additional space on the right for the text

# Display the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.savefig('PaperFigures/CancerPlot_2.png', dpi=500)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Function to create a horizontal bar plot
def create_plot(data, title, is_small_plot=False, ax=None, max_x_lim=None, remove_x_axis=False, remove_y_axis=False, remove_x_label=False, title_flag=''):
    if ax is None:
        fig, ax = plt.subplots(figsize=(3, 1) if is_small_plot else (3, 6))
        fig.patch.set_facecolor('none')
        ax.patch.set_facecolor('none')

    ax.grid(False)
    cancer_types, uncensored_counts, censored_counts = zip(*data)
    positions = np.arange(len(cancer_types))
    color_uncensored, color_censored = ('#ACD6FF', '#CECEFF') if 'classifier' not in title.lower() else ('#FFCBB3', '#FFF0AC')

    bar_width = 0.4
    ax.barh(positions, uncensored_counts, height=bar_width, edgecolor='white', color=color_uncensored, label='Censored' if 'classifier' not in title.lower() else 'High Risk')
    ax.barh(positions, censored_counts, left=uncensored_counts, height=bar_width, edgecolor='white', color=color_censored, label='Uncensored' if 'classifier' not in title.lower() else 'Low Risk')

    Uni_font_size = 11
    ax.tick_params(axis='x', labelsize=Uni_font_size)
    ax.tick_params(axis='y', labelsize=Uni_font_size)
    ax.set_yticks(positions)
    ax.set_yticklabels(cancer_types, fontsize=Uni_font_size)
    ax.set_xlabel('Number of patients', fontsize=Uni_font_size)
    if title_flag == 'a':
        ax.set_ylabel('Cancer Type', fontsize=Uni_font_size)
    else:
        ax.set_ylabel('', fontsize=Uni_font_size)

    max_total_count = max(unc + cen for unc, cen in zip(uncensored_counts, censored_counts))
    for pos, uncen, cen in zip(positions, uncensored_counts, censored_counts):
        total_count = f'({uncen}/{cen})'
        total_count_num = uncen + cen
        if total_count_num > 0:
            ax.text(total_count_num + 0.5, pos, total_count, va='center', fontsize=Uni_font_size-2)

    ax.set_xlim([0, max_total_count + (300 if not is_small_plot else 120)] if max_x_lim is None else [0, max_x_lim])
    if title_flag == 'a':
        title_get = ax.set_title('b. TCGA Training/Testing samples for Cox models (left)/Classifiers (right)', fontsize=Uni_font_size+1)
        title_get.set_position([1.3, 1])
    if title_flag == 'b':
        title_get = ax.set_title('c. Validation samples for Cox models (left)/Classifiers (right)', fontsize=Uni_font_size+1)
        title_get.set_position([1, 1])
    if 'validation samples for classifier' in title.lower():
        for pos in positions:
            if pos == 2:
                ax.text(max_x_lim * 0.9, pos, 'DKFZ', va='center', fontsize=Uni_font_size-2)
            else:
                ax.text(max_x_lim * 0.9, pos, 'CPTAC-3', va='center', fontsize=Uni_font_size-2)
    ax.figure.subplots_adjust(top=0.5)

    #original_ylim = ax.get_ylim()
    #print("Original Y-axis limits:", original_ylim)

    if not is_small_plot:
        ax.set_ylim([-0.3, 18.3])

    # Remove the spines (axis borders)
    for spine in ax.spines.values():
        spine.set_visible(False)

    if remove_x_axis:
        ax.set_xlabel('')
        ax.set_xticklabels([])
        ax.set_xticks([])
    else:
        ax.xaxis.set_tick_params(width=1, length=5, color='black')
        ax.axhline(-0.5, color='black', linewidth=1)
        for tick in ax.get_xticks():
            ax.plot(tick, -0.5, 's', color='black', markersize=3)  # 'ko' 表示黑色圆点

    if remove_y_axis:
        ax.set_ylabel('')
        ax.set_yticklabels([])
        ax.set_yticks([])

    if not remove_x_label:
        ax.legend(fontsize=Uni_font_size-2, loc='upper left', bbox_to_anchor=(0.75, 0.92))

    plt.tight_layout()
    return ax

# Function to combine plots into a single figure
def combine_plots_max_scale(data_dict):
    max_x_lim = 1100
    max_x_lim_risk = 550

    fig, axs = plt.subplots(2, 2, figsize=(6, 6), gridspec_kw={'height_ratios': [22, 3]})
    create_plot(data_dict['data_count'], 'TCGA samples for Cox model', ax=axs[0, 0], max_x_lim=max_x_lim, remove_x_axis=True, title_flag='a')
    create_plot(data_dict['data_count_risk'], 'TCGA samples for classifier', ax=axs[0, 1], max_x_lim=max_x_lim_risk, remove_x_axis=True, remove_y_axis=True)
    create_plot(data_dict['data_count_valid'], 'Validation samples for Cox model', is_small_plot=True, ax=axs[1, 0], max_x_lim=max_x_lim, remove_x_label=True, title_flag='b')
    create_plot(data_dict['data_count_risk_valid'], 'Validation samples for classifier', is_small_plot=True, ax=axs[1, 1], max_x_lim=max_x_lim_risk, remove_y_axis=True, remove_x_label=True)

    plt.tight_layout()
    plt.subplots_adjust(top=0.96, bottom=0.08)
    plt.savefig('PaperFigures/CancerPlots_Combined_Max_Scale.png', dpi=500, transparent=True)
    plt.savefig('PaperFigures/CancerPlots_Combined_Max_Scale.eps', format='eps', transparent=True)
    plt.show()
    return 'PaperFigures/CancerPlots_Combined_Max_Scale.png'

# Reading data from text files and storing in dictionary
file_paths = {
    'data_count': "PaperFigures/data_count.txt",
    'data_count_valid': "PaperFigures/data_count_valid.txt",
    'data_count_risk': "PaperFigures/data_count_risk.txt",
    'data_count_risk_valid': "PaperFigures/data_count_risk_valid.txt"
}

data = {key: [eval(line.strip()) for line in open(file_path)] for key, file_path in file_paths.items()}

# Combine plots with x-axis scale set to the longest value across all datasets, fixed bar height, and consistent bar height across plots
combined_figure_path_max_scale_fixed_height_consistent = combine_plots_max_scale(data)
combined_figure_path_max_scale_fixed_height_consistent
```
